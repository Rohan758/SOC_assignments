```python
def diagonallyDominant(arr):
     n = min(arr.shape)
     for i in range(n):
        row = arr[i, :]
        max_index = np.argmax(np.abs(row))
        if max_index !=i:
            arr[i,i], arr[i,max_index] = arr[i,max_index], arr[i,i]
        print(arr)

     return arr

arr= np.array([
    
 [  90 ,  69 ,   0 ,  -1 ,  -3],
 [   4 ,   2 ,   3 ,   7 , 406],
 [  11 , -12 ,  -3 ,  43 ,  33],
 [   7 ,   7 ,  -1 ,  -9 , -100],
 [ 100 ,  62 ,   2 ,   1 ,  11]
    
])

print("Original matrix:")
print(arr)

diagonallyDominant(arr)

print("Modified matrix:")
print(arr)




```

```python
row_sum = 0
cnt = 0
A = arr
problematic_rows = []

for i in range(n):
    de = abs(A[i,i])
    nde = np.sum(np.abs(A[i, :])) - de
    if de > nde:
        cnt +=1
    elif de < nde:
        cnt = 0
        problematic_rows.append((i, A[i, :]))
        break

if problematic_rows:
        print(f"{A} \n is not diagonally dominant")
        for row_index, row_elements in problematic_rows:
            print(f"Row {row_index} creates the problem: {row_elements}")

if cnt > 0:
    print(f"{A} is diagonally dominant")
```

```python
from matplotlib import pyplot as plt
import math

x = [i for i in range (-100,100)]
y_relu = []
y_sigmoid = []
y_tanh = []

for i in range (-100,100):
    y_relu.append(max(0, i))
    y_sigmoid.append(1 / (1 + math.exp(-i)))
    y_tanh.append(math.tanh(i))

plt.figure(figsize=(10,14))
plt.subplot(2,1, 1)
plt.plot(x, y_relu, color='green')

plt.subplot(2,1,2)
plt.plot(x,y_sigmoid, color='blue')
plt.plot(x,y_tanh, color='red')

plt.show()

```

```python
def softmax(arr):
    e_arr = np.exp(arr - np.max(arr))
    return e_arr / e_arr.sum()

arr = [90, 406, 43, -100, 100]
b = softmax(arr)
print("Softmax of the diagonal elements of the matrix :\n", b)
```

```python
plt.plot(softmax(arr), arr, 'o')
plt.show()
```
